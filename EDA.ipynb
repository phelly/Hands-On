##  Detecting The Outliers

#### using Z-Score

While calculating the Zscore we rescale and center the data and look for data points which are too far from zero. These data points which are way too far from zero will be treated as the outliers. In most of the cases a threshold of 3 or -3 is used ie if the Z-score value is greater than or less than 3 or -3 respectively,that data point will be identified as outliers. 

```python
from scipy import stats
```

```python
z=np.abs(stats.zscore(cars_df['Price']))
z
```

```python
threshold=3
print(np.where(z>3))
```

## Drawing a scatter plot

```python
#plotting a scatter plot
fig,ax = plt.subplots(figsize=(10,6))
ax.scatter(cars_df["Engine"],cars_df["Price"])
ax.set_ylabel("Engine")
ax.set_xlabel("Price")
plt.show()
```

```python
fig,ax = plt.subplots(figsize=(10,6))
ax.scatter(cars_df["MPG-C"],cars_df["Price"])
ax.set_ylabel("MPG-C")
ax.set_xlabel("Price")
plt.show()
```

```python
#plotting a histogram
cars_df.Make.value_counts().nlargest(40).plot(kind="bar",figsize=(10,5))
plt.title("Number of cars by Make")
plt.ylabel("Number of cars")
plt.xlabel("Make")
```

We can conclude that Chevrolet make has the highest number of cars as compared to Lotus Make
