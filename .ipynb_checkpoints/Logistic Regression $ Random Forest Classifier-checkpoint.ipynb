## Shannon Entropy

```python
def balance_calc(data, unit='natural'):
    base = {
        'shannon' : 2.,
        'natural' : math.exp(1),
        'hartley' : 10.
    }
    if len(data) <= 1:
        return 0
    counts = Counter()
    
    for d in data:
        counts[d] += 1
        
        ent = 0
        probs = [float(c)/ len(data) for c in counts.values()]
        for p in probs:
            if p > 0.:
                ent -= p * math.log(p, base[unit])
                
                return ent/math.log(len(data))
```

```python
balance_calc(data_df['BASKET_SIZE'],'shannon')
```

```python
# Feature extraction 
model_data_df = data_df[['SPEND','PROD_CODE','PROD_CODE_40','SHOP_WEEKDAY','SHOP_HOUR','QUANTITY','CUST_LIFESTAGE','BASKET_SIZE','CUST_PRICE_SENSITIVITY']]
model_data_df.head()
```

```python
#extraction the independent and dependent variables
y = model_data_df['SPEND']
x = model_data_df[['PROD_CODE','PROD_CODE_40','SHOP_WEEKDAY','SHOP_HOUR','QUANTITY','BASKET_SIZE','CUST_LIFESTAGE','CUST_PRICE_SENSITIVITY']]
```

```python
x.head()
```

```python
x.isnull().sum()
```

```python
x.fillna('Unknown',inplace=True)
```

```python
x.isnull().sum()
```

# Encoding

```python
#frequency encoding
CUST_LIFESTAGE_fe = x.groupby('CUST_LIFESTAGE').size()/len(x)
x.loc[:,'CUST_LIFESTAGE_FREQ_ENC'] = x['CUST_LIFESTAGE'].map(CUST_LIFESTAGE_fe)
```

```python
prod_code_fe = x.groupby('PROD_CODE').size()/len(x)
x.loc[:,'PROD_CODE_FREQ'] = x['PROD_CODE'].map(prod_code_fe)
```

```python
basket_size_fe = x.groupby('BASKET_SIZE').size()/len(x)
x.loc[:,'BASKET_SIZE_FREQ'] = x['BASKET_SIZE'].map(basket_size_fe)
```

```python
cust_price_sensitivity_fe = x.groupby('CUST_PRICE_SENSITIVITY').size()/len(x)
x.loc[:,'CUST_PRICE_SENSITIVITY_FREQ'] = x['CUST_PRICE_SENSITIVITY'].map(cust_price_sensitivity_fe)
```

```python
x.head()
```

```python
prod_code_40 =x.groupby('PROD_CODE_40').size()/len(x)
x.loc[:,'PROD_CODE_40_FREQ'] = x['PROD_CODE_40'].map(prod_code_40)
```

```python
x.head()
```

```python
#drop irrelevant columns
x.drop(columns=['PROD_CODE'],inplace=True)
x.drop(columns=['PROD_CODE_40'],inplace = True)
x.drop(columns = ['BASKET_SIZE'],inplace = True)
x.drop(columns = ['CUST_LIFESTAGE'], inplace = True)
x.drop(columns = ['CUST_PRICE_SENSITIVITY'],inplace = True)
```

```python
#splitting the datatsets in training and test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
logmodel = LogisticRegression()
logmodel.fit(x_train,y_train)
```

```python
predictions = logmodel.predict(x_test)
predictions
```

```python
#model accuracy
print('Accuracy:',metrics.accuracy_score(y_test,predictions))
```
